```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Exponential
exp_data = np.random.exponential(scale=1.0, size=100000)
```

```python
plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue",
label="rand")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

 3D plot

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("Z = X² + Y²")
plt.show()
```

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
```

```python
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
df.head()
```

```python
features = ['Hp', 'Attack', 'Defense', 'Sp_attack', 'Sp_defense', 'Speed']
```

```python
corr_pearson = df[features].corr(method='pearson')
```

```python
print(df.columns)
```

```python
# Pearson Correlation
corr_pearson = df[features].corr(method='pearson')

# Spearman Correlation
corr_spearman = df[features].corr(method='spearman')
```

```python
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
```

```python
corr_pearson = df[features].corr(method='pearson')
corr_spearman = df[features].corr(method='spearman')
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(corr_pearson, annot=True, cmap='coolwarm')
plt.title("Pearson Correlation Heatmap")
plt.show()
```

```python

plt.figure(figsize=(10, 5))
sns.heatmap(corr_spearman, annot=True, cmap='coolwarm')
plt.title("Spearman Correlation Heatmap")
plt.show()
```
